## Q1 - Manipulando o DataFrame

Crie uma coluna chamada `Hemisfério` baseada na Latitude

A regra de formação é a seguinte:

Valor | Critério
---|---
Norte | Latitude positiva
Sul | Latitude negativa

```python
Hemisferio = []

for k in df['Latitude']:
    if k<0:
        Hemisferio.append('Sul')
    else:
        Hemisferio.append('Norte')

df['Hemisfério'] = Hemisferio
#df
```

## Q6 - Tirando conclusões com base nos dados

Em um certo lugar já ocorreram abalos com *Magnitude Type* `MB` e *Type* `Nuclear Explosion`. 
 
Responda: 
    * É mais provável que tenha sido no norte ou no sul? 
    
Assuma que os Magnitude Type e Type são independentes


```python
df2 = df[df['Magnitude Type'].isin(['MB'])]
tabela = pd.crosstab(df2.Hemisfério,df["Magnitude Type"], margins=True, dropna=True, normalize=True)*100
tabela
```

    Como podemos ver, a crosstab mostra que a porcentagem de ocorrer o MB no Sul é maior que a do Norte. Por isso, é mais provável que tenha sido no Sul.
